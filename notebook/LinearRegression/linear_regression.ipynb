```python
# build data for the line assume (x,y) is around the line 7x+5  , 100 point
import numpy as np
import matplotlib.pyplot as plt
# np.random.seed(45)
num_of_point = 100
X = 2*np.random.rand(num_of_point,1) # build X from [0,5]
Y = 5 + 7*X + np.random.randn(num_of_point,1) # build Y using X and + bias equal to 5
# visualize the dataset
plt.plot(X,Y,'b.')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0,2,0,25])
plt.show()
```

```python
def MSE(theta) :
    X_1 = np.c_[X,np.ones((m,1))]
    return 1/X.shape[0] * np.sum((X_1.dot(theta)-Y)**2)
```

```python
# prepare data by add column of full 1 to X
m = X.shape[0] # number of point
ones = np.ones((m,1))
X_1 = np.c_[X,ones]
# X = X_1

# batch gradient descent
eta = 0.1
theta = np.random.rand(2,1)
iteration = 10000
while iteration!=0 :
    gradient = 2/m * X_1.T.dot(X_1.dot(theta)-Y)
    theta = theta - eta*gradient
    iteration-=1

plt.plot(X,Y,'b.')
theta_best = np.linalg.inv(X_1.T.dot(X_1)).dot(X_1.T).dot(Y)
X_new = np.array([[0],[2]])
Y_new = np.c_[X_new,np.ones([2,1])].dot(theta_best)
plt.plot(X_new,Y_new,'g-',label='best fit line',linewidth = 5)
plt.plot([0,2],[theta[1][0],theta[1][0]+2*theta[0][0]],'r-',label = 'Batch gradient descent')
plt.plot
plt.legend(loc = 'upper left' , fontsize = 12)
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$y$", rotation=0, fontsize=12)
plt.axis([0,2,0,25])
plt.show()
print("min error " ,  MSE(theta_best))
print("my error  " , MSE(theta))
```
