Linear Regression

Build dataset by numpy.random

```python
import numpy as np 

X = 2*np.random.rand(100,1)  
Y = 4+ 3*X + 3*np.random.rand(100,1)
```

```python
plt.plot(X,Y,'b.')
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation = 0  , fontsize=18)
plt.axis([0,2,0,15])
plt.show()
```
